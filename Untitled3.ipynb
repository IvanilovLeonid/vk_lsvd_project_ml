```python
import polars as pl
import numpy as np

train_interactions = pl.read_parquet('processed/train_interactions.parquet')
internal_val_interactions = pl.read_parquet('processed/internal_val_interactions.parquet')
test_interactions = pl.read_parquet('processed/test_interactions.parquet')
items_metadata = pl.read_parquet('processed/items_metadata.parquet')
```

```python
# Веса подобраны по силе сигнала: явные позитивные действия весят больше,
# dislike штрафует, watch_ratio даёт непрерывную базу
W_LIKE = 3.0
W_DISLIKE = -4.0
W_SHARE = 4.0
W_BOOKMARK = 3.5
W_CLICK_AUTHOR = 1.0
W_COMMENTS = 1.5

def add_score(df: pl.DataFrame, items_meta: pl.DataFrame) -> pl.DataFrame:
    df = df.join(items_meta.select(['item_id', 'duration']), on='item_id', how='left')
    df = df.with_columns([
        (pl.col('duration').fill_null(1).clip(lower_bound=1)).alias('duration_safe')
    ])
    df = df.with_columns([
        (pl.col('timespent') / pl.col('duration_safe')).clip(upper_bound=1.0).alias('watch_ratio')
    ])
    df = df.with_columns([
        (
            pl.col('watch_ratio')
            + pl.col('like').cast(pl.Float32) * W_LIKE
            + pl.col('dislike').cast(pl.Float32) * W_DISLIKE
            + pl.col('share').cast(pl.Float32) * W_SHARE
            + pl.col('bookmark').cast(pl.Float32) * W_BOOKMARK
            + pl.col('click_on_author').cast(pl.Float32) * W_CLICK_AUTHOR
            + pl.col('open_comments').cast(pl.Float32) * W_COMMENTS
        ).alias('score')
    ])

    df = df.with_columns(pl.col('score').clip(lower_bound=0.0))
    return df

train_scored = add_score(train_interactions, items_metadata)
print(train_scored.select(['user_id', 'item_id', 'watch_ratio', 'score']).describe())
```

```python

def add_relevance(df: pl.DataFrame, items_meta: pl.DataFrame) -> pl.DataFrame:
    df = df.join(items_meta.select(['item_id', 'duration']), on='item_id', how='left')
    df = df.with_columns(pl.col('duration').fill_null(1).clip(lower_bound=1).alias('duration_safe'))
    df = df.with_columns((pl.col('timespent') / pl.col('duration_safe')).alias('watch_ratio'))
    df = df.with_columns(
        (
            (pl.col('watch_ratio') >= 0.5) |
            pl.col('like') | pl.col('share') | pl.col('bookmark')
        ).alias('relevant')
    )
    return df.filter(pl.col('relevant'))

train_relevant = add_relevance(train_interactions, items_metadata)
val_relevant = add_relevance(internal_val_interactions, items_metadata)
test_relevant = add_relevance(test_interactions, items_metadata)

print('Позитивов в train:', train_relevant.height, '/', train_interactions.height)
print('Позитивов в internal val:', val_relevant.height)
print('Позитивов в test:', test_relevant.height)
```

```python
def build_ground_truth(relevant_df: pl.DataFrame):
    df = relevant_df.filter(
        pl.col('user_id').is_in(train_user_ids) & pl.col('item_id').is_in(train_item_ids)
    )
    df = df.with_columns([
        pl.col('user_id').replace_strict(user_id_to_idx, return_dtype=pl.Int64).alias('u_idx'),
        pl.col('item_id').replace_strict(item_id_to_idx, return_dtype=pl.Int64).alias('i_idx'),
    ])
    gt = df.group_by('u_idx').agg(pl.col('i_idx').unique().alias('relevant_items'))
    return dict(zip(gt['u_idx'].to_list(), gt['relevant_items'].to_list()))

val_ground_truth = build_ground_truth(val_relevant)
test_ground_truth = build_ground_truth(test_relevant)

print(f'Пользователей с ground truth в val: {len(val_ground_truth)}')
print(f'Пользователей с ground truth в test: {len(test_ground_truth)}')
```

```python
import numpy as np

def recall_at_k(recommended: list, relevant: set, k: int) -> float:
    if len(relevant) == 0:
        return np.nan
    rec_k = set(recommended[:k])
    return len(rec_k & relevant) / len(relevant)

def precision_at_k(recommended: list, relevant: set, k: int) -> float:
    rec_k = recommended[:k]
    if len(rec_k) == 0:
        return np.nan
    return len(set(rec_k) & relevant) / len(rec_k)

def ndcg_at_k(recommended: list, relevant: set, k: int) -> float:
    rec_k = recommended[:k]
    dcg = sum(1.0 / np.log2(i + 2) for i, item in enumerate(rec_k) if item in relevant)
    idcg = sum(1.0 / np.log2(i + 2) for i in range(min(len(relevant), k)))
    if idcg == 0:
        return np.nan
    return dcg / idcg

def average_precision_at_k(recommended: list, relevant: set, k: int) -> float:
    rec_k = recommended[:k]
    hits, ap = 0, 0.0
    for i, item in enumerate(rec_k):
        if item in relevant:
            hits += 1
            ap += hits / (i + 1)
    if len(relevant) == 0:
        return np.nan
    return ap / min(len(relevant), k)

def evaluate_recommendations(recs_by_user: dict, ground_truth: dict, k: int = 10) -> dict:
    """recs_by_user: {u_idx: [item_idx, ...]}, ground_truth: {u_idx: [item_idx, ...]}"""
    recalls, precisions, ndcgs, maps = [], [], [], []
    recommended_items_union = set()
    for u_idx, relevant_items in ground_truth.items():
        relevant_set = set(relevant_items)
        recommended = recs_by_user.get(u_idx, [])
        recommended_items_union.update(recommended[:k])
        recalls.append(recall_at_k(recommended, relevant_set, k))
        precisions.append(precision_at_k(recommended, relevant_set, k))
        ndcgs.append(ndcg_at_k(recommended, relevant_set, k))
        maps.append(average_precision_at_k(recommended, relevant_set, k))
    return {
        'recall@k': np.nanmean(recalls),
        'precision@k': np.nanmean(precisions),
        'ndcg@k': np.nanmean(ndcgs),
        'map@k': np.nanmean(maps),
        'coverage': len(recommended_items_union) / n_items,
    }
```

```python
from sklearn.preprocessing import normalize
from scipy.sparse import csr_matrix, vstack
import numpy as np

def build_item_similarity_topn(train_matrix: csr_matrix, top_sim: int = 200, chunk_size: int = 500) -> csr_matrix:
    item_matrix = normalize(train_matrix.T, norm='l2', axis=1).tocsr()  # (n_items, n_users)
    n_items_local = item_matrix.shape[0]
    item_matrix_T = item_matrix.T.tocsr()

    blocks = []
    for start in range(0, n_items_local, chunk_size):
        end = min(start + chunk_size, n_items_local)
        chunk = item_matrix[start:end]
        sims_dense = (chunk @ item_matrix_T).toarray()

        if sims_dense.shape[1] > top_sim:
            idx_part = np.argpartition(-sims_dense, top_sim, axis=1)[:, :top_sim]
            mask = np.zeros_like(sims_dense, dtype=bool)
            np.put_along_axis(mask, idx_part, True, axis=1)
            sims_dense[~mask] = 0.0

        blocks.append(csr_matrix(sims_dense))
        if (start // chunk_size) % 10 == 0:
            print(f'  processed items {start}-{end}/{n_items_local}')

    return vstack(blocks).tocsr()

def build_item_knn_recs(train_matrix: csr_matrix, users_to_score: list, k: int = 10, top_sim: int = 200) -> dict:
    sim_matrix = build_item_similarity_topn(train_matrix, top_sim=top_sim, chunk_size=500)

    recs = {}
    train_matrix_csr = train_matrix.tocsr()
    for u_idx in users_to_score:
        user_row = train_matrix_csr.getrow(u_idx)
        seen_items = set(user_row.indices)
        if len(seen_items) == 0:
            recs[u_idx] = []
            continue
        scores = (user_row @ sim_matrix).toarray().flatten()
        scores[list(seen_items)] = -np.inf
        top_items = np.argpartition(-scores, k)[:k]
        top_items = top_items[np.argsort(-scores[top_items])]
        recs[u_idx] = top_items.tolist()
    return recs

itemknn_recs = build_item_knn_recs(train_matrix, list(val_ground_truth.keys()), k=TOP_K)
itemknn_metrics_val = evaluate_recommendations(itemknn_recs, val_ground_truth, k=TOP_K)
print('ItemKNN (val):', itemknn_metrics_val)
```

```python
from implicit.als import AlternatingLeastSquares

def train_als(train_matrix, seed, factors=64, regularization=0.05, iterations=15):
    model = AlternatingLeastSquares(
        factors=factors, regularization=regularization,
        iterations=iterations, random_state=seed
    )
    model.fit(train_matrix)
    return model

def get_als_recs(model, train_matrix, users_to_score: list, k: int = 10) -> dict:
    user_items = train_matrix.tocsr()
    ids, scores = model.recommend(
        users_to_score, user_items[users_to_score], N=k, filter_already_liked_items=True
    )
    return {u: list(ids[i]) for i, u in enumerate(users_to_score)}

SEEDS = [42, 123, 2026]
als_metrics_per_seed = []
als_models = {}

val_users = list(val_ground_truth.keys())
for seed in SEEDS:
    model = train_als(train_matrix, seed=seed)
    als_models[seed] = model
    recs = get_als_recs(model, train_matrix, val_users, k=TOP_K)
    metrics = evaluate_recommendations(recs, val_ground_truth, k=TOP_K)
    metrics['seed'] = seed
    als_metrics_per_seed.append(metrics)
    print(f'ALS seed={seed}:', metrics)

import pandas as pd
als_df = pd.DataFrame(als_metrics_per_seed)
print('\nALS mean ± std по сидам:')
print(als_df[['recall@k', 'ndcg@k', 'map@k', 'coverage']].agg(['mean', 'std']))
```

столкнулись с проблемой при установке lightfm, найдем аналог

Идея модели: BPR (Bayesian Personalized Ranking) — тот же принцип ранжирования, что и в LightFM с loss='warp'-подобным поведением

```python
import torch
import torch.nn as nn
import numpy as np

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('Device:', device)

class HybridBPR(nn.Module):
    def __init__(self, n_users, n_items, factors=64, content_dim=None):
        super().__init__()
        self.user_emb = nn.Embedding(n_users, factors)
        self.item_emb = nn.Embedding(n_items, factors)
        nn.init.normal_(self.user_emb.weight, std=0.05)
        nn.init.normal_(self.item_emb.weight, std=0.05)
        self.content_dim = content_dim
        if content_dim is not None:
            self.content_proj = nn.Linear(content_dim, factors, bias=False)
            nn.init.normal_(self.content_proj.weight, std=0.05)

    def item_representation(self, content_emb=None):
        base = self.item_emb.weight  # (n_items, factors)
        if self.content_dim is not None and content_emb is not None:
            base = base + self.content_proj(content_emb)
        return base

    def forward(self, u_idx, pos_idx, neg_idx, content_emb=None):
        item_repr = self.item_representation(content_emb)
        u = self.user_emb(u_idx)
        pos_score = (u * item_repr[pos_idx]).sum(dim=1)
        neg_score = (u * item_repr[neg_idx]).sum(dim=1)
        return pos_score, neg_score


def build_positive_pairs(train_scored, user_id_to_idx, item_id_to_idx):
    df = train_scored.filter(pl.col('score') > 0)
    df = df.with_columns([
        pl.col('user_id').replace_strict(user_id_to_idx, return_dtype=pl.Int64).alias('u_idx'),
        pl.col('item_id').replace_strict(item_id_to_idx, return_dtype=pl.Int64).alias('i_idx'),
    ])
    return df.select(['u_idx', 'i_idx']).to_numpy()

positive_pairs = build_positive_pairs(train_scored, user_id_to_idx, item_id_to_idx)
print('Позитивных пар для BPR:', positive_pairs.shape[0])

def train_hybrid_bpr(positive_pairs, n_users, n_items, seed, content_dim=None,
                      content_emb_np=None, factors=64, epochs=18, batch_size=8192, lr=0.01):
    torch.manual_seed(seed)
    np.random.seed(seed)
    model = HybridBPR(n_users, n_items, factors=factors, content_dim=content_dim).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=1e-6)

    content_tensor = None
    if content_emb_np is not None:
        content_tensor = torch.tensor(content_emb_np, dtype=torch.float32, device=device)

    u_all = torch.tensor(positive_pairs[:, 0], dtype=torch.long, device=device)
    i_all = torch.tensor(positive_pairs[:, 1], dtype=torch.long, device=device)
    n_pairs = len(u_all)

    for epoch in range(epochs):
        perm = torch.randperm(n_pairs, device=device)
        total_loss = 0.0
        n_batches = 0
        for start in range(0, n_pairs, batch_size):
            batch_idx = perm[start:start + batch_size]
            u_b = u_all[batch_idx]
            pos_b = i_all[batch_idx]
            neg_b = torch.randint(0, n_items, (len(batch_idx),), device=device)

            pos_score, neg_score = model(u_b, pos_b, neg_b, content_tensor)
            loss = -torch.log(torch.sigmoid(pos_score - neg_score) + 1e-8).mean()

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            n_batches += 1
        print(f'  epoch {epoch+1}/{epochs}, loss={total_loss/n_batches:.4f}')
    return model, content_tensor


def get_hybrid_recs(model, content_tensor, train_matrix, users_to_score, k=10):
    model.eval()
    with torch.no_grad():
        item_repr = model.item_representation(content_tensor)  # (n_items, factors)
        train_csr = train_matrix.tocsr()
        recs = {}
        u_idx_tensor = torch.tensor(users_to_score, dtype=torch.long, device=device)
        user_vecs = model.user_emb(u_idx_tensor)  # (n_users_to_score, factors)
        scores_all = (user_vecs @ item_repr.T).cpu().numpy()  # (n_users_to_score, n_items)
        for row, u_idx in enumerate(users_to_score):
            seen = set(train_csr.getrow(u_idx).indices)
            scores = scores_all[row].copy()
            if seen:
                scores[list(seen)] = -np.inf
            top_items = np.argpartition(-scores, k)[:k]
            top_items = top_items[np.argsort(-scores[top_items])]
            recs[u_idx] = top_items.tolist()
    return recs


hybrid_configs = {'cf_only': None, 'emb_8': 8, 'emb_32': 32, 'emb_64': 64}
hybrid_results = {}

for config_name, dim in hybrid_configs.items():
    print(f'\n=== Hybrid BPR: {config_name} ===')
    content_emb_np = get_embeddings(dim) if dim is not None else None
    metrics_per_seed = []
    for seed in SEEDS:
        model, content_tensor = train_hybrid_bpr(
            positive_pairs, n_users, n_items, seed=seed,
            content_dim=dim, content_emb_np=content_emb_np
        )
        recs = get_hybrid_recs(model, content_tensor, train_matrix, val_users, k=TOP_K)
        m = evaluate_recommendations(recs, val_ground_truth, k=TOP_K)
        m['seed'] = seed
        metrics_per_seed.append(m)
        print(f'  seed={seed}: recall={m["recall@k"]:.4f} ndcg={m["ndcg@k"]:.4f}')
    hybrid_results[config_name] = metrics_per_seed
```

```python
def rerank_by_content(als_recs_pool: dict, user_history_matrix, embeddings, alpha=0.5, k=10, pool_size=50):
    """Берём top-pool_size от ALS, ре-ранжируем как alpha*als_score + (1-alpha)*content_sim к истории юзера."""
    emb_norm = embeddings / (np.linalg.norm(embeddings, axis=1, keepdims=True) + 1e-8)
    reranked = {}
    train_csr = user_history_matrix.tocsr()
    for u_idx, candidates in als_recs_pool.items():
        seen_items = train_csr.getrow(u_idx).indices
        if len(seen_items) == 0 or len(candidates) == 0:
            reranked[u_idx] = candidates[:k]
            continue
        user_profile = emb_norm[seen_items].mean(axis=0, keepdims=True)
        cand_emb = emb_norm[candidates]
        content_sim = (cand_emb @ user_profile.T).flatten()
        als_rank_score = np.linspace(1, 0, len(candidates))  # позиционный скор от ALS-порядка
        combined = alpha * als_rank_score + (1 - alpha) * content_sim
        order = np.argsort(-combined)
        reranked[u_idx] = [candidates[i] for i in order[:k]]
    return reranked

# получаем расширенный пул кандидатов от ALS (seed=42) для ре-ранка
als_model_42 = als_models[42]
als_pool_recs = get_als_recs(als_model_42, train_matrix, val_users, k=50)

rerank_results = {}
for dim in [8, 32, 64]:
    embeddings = get_embeddings(dim)
    reranked = rerank_by_content(als_pool_recs, train_matrix, embeddings, alpha=0.5, k=TOP_K)
    m = evaluate_recommendations(reranked, val_ground_truth, k=TOP_K)
    rerank_results[f'als_rerank_emb_{dim}'] = m
    print(f'ALS + rerank (dim={dim}):', m)

```
